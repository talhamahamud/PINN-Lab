## 1. What is an ODE?

An **Ordinary Differential Equation (ODE)** is an equation involving:
- An **unknown function** $u(t)$ (or $y(x)$)
- Its **derivatives** with respect to a **single** independent variable

### General Form:
$$F\left(t,\ u,\ u',\ u'',\ \ldots,\ u^{(n)}\right) = 0$$

where $u^{(n)} = \dfrac{d^n u}{dt^n}$

### Simple Example:
$$\frac{du}{dt} = -k \cdot u$$

This says: *"The rate of change of $u$ is proportional to $u$ itself."*  
This models radioactive decay, cooling, etc.

### ODE vs PDE:
| | ODE | PDE |
|---|---|---|
| Independent variables | One (e.g., $t$) | Multiple (e.g., $x, t$) |
| Derivatives | Ordinary $\frac{du}{dt}$ | Partial $\frac{\partial u}{\partial t}$, $\frac{\partial u}{\partial x}$ |
| Example | $u' = -ku$ | $\frac{\partial u}{\partial t} = \alpha \frac{\partial^2 u}{\partial x^2}$ |

> 💡 **PINNs can solve both ODEs and PDEs.** We start with ODEs to build intuition.

---
## 2. Key Terminology

| Term | Meaning | Example |
|---|---|---|
| **Order** | Highest derivative in the equation | $u'' + u = 0$ → 2nd order |
| **Degree** | Power of the highest derivative | $(u')^3 + u = 0$ → degree 3 |
| **Linear** | $u$ and its derivatives appear to the 1st power only | $u'' + 3u' + 2u = 0$ |
| **Nonlinear** | $u$ or derivatives appear with powers > 1 or in products | $u' = u^2$, $u'' \cdot u' = t$ |
| **Homogeneous** | No term independent of $u$ | $u'' + u = 0$ |
| **Non-homogeneous** | Has a forcing term | $u'' + u = \sin(t)$ |
| **General Solution** | Family of all solutions (includes constants $C_1, C_2, ...$) | $u = Ce^{-kt}$ |
| **Particular Solution** | Specific solution after applying conditions | $u = 5e^{-2t}$ |

## 3. Types of ODEs

```
ODEs
├── Linear
│   ├── 1st Order
│   │   ├── Separable
│   │   └── Integrating Factor
│   └── 2nd Order (constant coefficients)
│       ├── Homogeneous
│       └── Non-homogeneous (Undetermined Coefficients, Variation of Parameters)
└── Nonlinear
    ├── Bernoulli Equations
    ├── Exact Equations
    └── Systems of ODEs
```

For PINNs, we care most about **clearly stated ODEs with known boundary/initial conditions**, regardless of type.

---
## 4. IVP vs BVP

### Initial Value Problem (IVP)
Conditions are specified at a **single point** (usually $t = 0$):
$$u'' + u = 0, \quad u(0) = 1, \quad u'(0) = 0$$

- Think: *"I know the starting state, find the future."*
- Used in: time evolution problems (mechanics, circuits)

### Boundary Value Problem (BVP)
Conditions are specified at **multiple points** (usually endpoints of a domain):
$$u'' + u = 0, \quad u(0) = 0, \quad u(\pi) = 0$$

- Think: *"I know the state at the boundaries, find the solution in between."*
- Used in: structural mechanics, heat distribution

> 🔑 **PINNs handle both IVPs and BVPs naturally** — the initial/boundary conditions become part of the loss function.

## 5. Analytical Solutions

Understanding analytical solutions is **crucial for PINNs** — we use them to verify our neural network's output.

---
### 5.1 Separable ODEs

**Form:** $\dfrac{du}{dt} = f(t) \cdot g(u)$

**Method:** Separate variables, integrate both sides.

$$\frac{du}{g(u)} = f(t)\, dt \implies \int \frac{du}{g(u)} = \int f(t)\, dt$$

**Example:** Exponential Decay
$$\frac{du}{dt} = -k u, \quad u(0) = u_0$$

**Solution steps:**
$$\frac{du}{u} = -k\, dt$$
$$\int \frac{du}{u} = \int -k\, dt$$
$$\ln|u| = -kt + C$$
$$u(t) = A e^{-kt}$$

Apply IC: $u(0) = u_0 \implies A = u_0$

$$\boxed{u(t) = u_0\, e^{-kt}}$$

---
### 5.2 First-Order Linear ODE (Integrating Factor)

**Form:** $\dfrac{du}{dt} + P(t)\, u = Q(t)$

**Method:** Multiply by integrating factor $\mu(t) = e^{\int P(t)dt}$

$$\mu(t)\, u(t) = \int \mu(t)\, Q(t)\, dt + C$$

**Example:**
$$\frac{du}{dt} + 2u = 4, \quad u(0) = 0$$

Here $P(t) = 2$, $Q(t) = 4$

$\mu = e^{2t}$

$$e^{2t} u = \int 4 e^{2t} dt = 2e^{2t} + C$$
$$u = 2 + Ce^{-2t}$$

Apply IC: $u(0) = 0 \implies 0 = 2 + C \implies C = -2$

$$\boxed{u(t) = 2(1 - e^{-2t})}$$

---
### 5.3 Second-Order Linear ODE with Constant Coefficients

**Form:** $a u'' + b u' + c u = 0$

**Method:** Assume $u = e^{\lambda t}$, substitute to get the **characteristic equation:**

$$a\lambda^2 + b\lambda + c = 0$$

**Three cases based on discriminant $\Delta = b^2 - 4ac$:**

| Case | Condition | Solution Form |
|---|---|---|
| Two real roots | $\Delta > 0$ | $u = C_1 e^{\lambda_1 t} + C_2 e^{\lambda_2 t}$ |
| Repeated root | $\Delta = 0$ | $u = (C_1 + C_2 t) e^{\lambda t}$ |
| Complex roots $\alpha \pm \beta i$ | $\Delta < 0$ | $u = e^{\alpha t}(C_1 \cos(\beta t) + C_2 \sin(\beta t))$ |

**Example (Simple Harmonic Oscillator):** $u'' + \omega^2 u = 0$

Characteristic equation: $\lambda^2 + \omega^2 = 0 \implies \lambda = \pm i\omega$

Complex roots: $\alpha = 0$, $\beta = \omega$

$$\boxed{u(t) = C_1 \cos(\omega t) + C_2 \sin(\omega t)}$$

## 6. Classic ODE Examples (with Plots)

Now let's implement and visualize each example numerically and analytically.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Global plot settings
plt.rcParams.update({
    'figure.dpi': 120,
    'axes.grid': True,
    'grid.alpha': 0.3,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'font.size': 12
})
print("Libraries loaded successfully!")
````

### 6.1 Exponential Decay

$$\frac{du}{dt} = -k\, u, \quad u(0) = u_0$$

**Analytical solution:** $u(t) = u_0\, e^{-kt}$

**Physical meaning:** Radioactive decay, cooling, drug concentration in blood

````python
# --- Parameters ---
k  = 0.5    # decay constant
u0 = 10.0   # initial value
t  = np.linspace(0, 10, 300)

# --- Analytical Solution ---
u_exact = u0 * np.exp(-k * t)

# --- Plot ---
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Solution curve
axes[0].plot(t, u_exact, 'royalblue', lw=2.5, label=r'$u(t) = u_0 e^{-kt}$')
axes[0].axhline(0, color='gray', linestyle='--', alpha=0.5)
axes[0].set_xlabel('Time $t$')
axes[0].set_ylabel('$u(t)$')
axes[0].set_title('Exponential Decay')
axes[0].legend()

# Effect of different k values
for ki, color in zip([0.2, 0.5, 1.0, 2.0], ['green', 'royalblue', 'orange', 'red']):
    axes[1].plot(t, u0 * np.exp(-ki * t), color=color, lw=2, label=f'k = {ki}')
axes[1].set_xlabel('Time $t$')
axes[1].set_ylabel('$u(t)$')
axes[1].set_title('Effect of Decay Constant $k$')
axes[1].legend()

plt.suptitle('Exponential Decay: $du/dt = -ku$', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Half-life
half_life = np.log(2) / k
print(f"Half-life (k={k}): t_½ = ln(2)/k = {half_life:.4f}")
print(f"At t = {half_life:.2f}, u = {u0 * np.exp(-k * half_life):.4f} ≈ u₀/2 = {u0/2}")
````

### 6.2 Simple Harmonic Oscillator (SHO)

$$u'' + \omega^2 u = 0, \quad u(0) = 1, \quad u'(0) = 0$$

**Analytical solution:** $u(t) = \cos(\omega t)$

**Physical meaning:** Ideal spring-mass system (no friction), pendulum for small angles

> 🎯 This is **the most important PINN benchmark problem.** We will solve it with a PINN in notebook `03_pinn_basics/01_simple_ode.ipynb`.

````python
# --- Parameters ---
omega = 2.0    # angular frequency
u0_sho = 1.0  # initial displacement
v0_sho = 0.0  # initial velocity
t = np.linspace(0, 4 * np.pi / omega, 500)

# --- Analytical Solution ---
# General: u(t) = C1*cos(wt) + C2*sin(wt)
# IC: u(0)=1 → C1=1; u'(0)=0 → C2=0
C1 = u0_sho
C2 = v0_sho / omega
u_sho = C1 * np.cos(omega * t) + C2 * np.sin(omega * t)
v_sho = -C1 * omega * np.sin(omega * t) + C2 * omega * np.cos(omega * t)

# --- Phase Portrait ---
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Time series
axes[0].plot(t, u_sho, 'royalblue', lw=2.5, label='displacement $u(t)$')
axes[0].plot(t, v_sho, 'tomato', lw=2, linestyle='--', label="velocity $u'(t)$")
axes[0].set_xlabel('Time $t$')
axes[0].set_ylabel('Amplitude')
axes[0].set_title('Simple Harmonic Oscillator')
axes[0].legend()

# Phase portrait (u vs u')
axes[1].plot(u_sho, v_sho, 'purple', lw=2.5)
axes[1].plot(u_sho[0], v_sho[0], 'go', markersize=8, label='Start')
axes[1].set_xlabel('Displacement $u$')
axes[1].set_ylabel("Velocity $u'$")
axes[1].set_title('Phase Portrait (closed orbit = no energy loss)')
axes[1].legend()
axes[1].set_aspect('equal')

plt.suptitle(r"Simple Harmonic Oscillator: $u'' + \omega^2 u = 0$", fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Energy check
E = 0.5 * v_sho**2 + 0.5 * omega**2 * u_sho**2
print(f"Total energy (should be constant): min={E.min():.6f}, max={E.max():.6f}")
````

### 6.3 Damped Harmonic Oscillator

$$u'' + 2\gamma u' + \omega_0^2 u = 0, \quad u(0) = 1, \quad u'(0) = 0$$

**Characteristic equation:** $\lambda^2 + 2\gamma\lambda + \omega_0^2 = 0$

**Three regimes based on $\gamma$ vs $\omega_0$:**

| Regime | Condition | Behavior |
|---|---|---|
| Underdamped | $\gamma < \omega_0$ | Oscillates, decaying amplitude |
| Critically damped | $\gamma = \omega_0$ | Fastest decay to zero, no oscillation |
| Overdamped | $\gamma > \omega_0$ | Slow exponential decay, no oscillation |

````python
omega0 = 2.0    # natural frequency
t = np.linspace(0, 10, 600)

def damped_oscillator(t, gamma, omega0=2.0, u0=1.0, v0=0.0):
    """Analytical solution of damped harmonic oscillator."""
    disc = gamma**2 - omega0**2  # discriminant / delta

    if disc < 0:  # Underdamped
        wd = np.sqrt(omega0**2 - gamma**2)  # damped frequency
        A = u0
        B = (v0 + gamma * u0) / wd
        return np.exp(-gamma * t) * (A * np.cos(wd * t) + B * np.sin(wd * t))

    elif disc == 0:  # Critically damped
        lam = -gamma
        C1 = u0
        C2 = v0 - lam * u0
        return (C1 + C2 * t) * np.exp(lam * t)

    else:  # Overdamped
        lam1 = -gamma + np.sqrt(disc)
        lam2 = -gamma - np.sqrt(disc)
        C2 = (v0 - lam1 * u0) / (lam2 - lam1)
        C1 = u0 - C2
        return C1 * np.exp(lam1 * t) + C2 * np.exp(lam2 * t)

# --- Three regimes ---
configs = [
    (0.3, 'Underdamped (γ=0.3 < ω₀=2)', 'royalblue'),
    (2.0, 'Critically Damped (γ=ω₀=2)', 'green'),
    (4.0, 'Overdamped (γ=4 > ω₀=2)', 'tomato'),
]

plt.figure(figsize=(11, 5))
for gamma, label, color in configs:
    u = damped_oscillator(t, gamma, omega0)
    plt.plot(t, u, color=color, lw=2.5, label=label)

plt.axhline(0, color='gray', linestyle=':', alpha=0.7)
plt.xlabel('Time $t$')
plt.ylabel('$u(t)$')
plt.title(r"Damped Harmonic Oscillator: $u'' + 2\gamma u' + \omega_0^2 u = 0$", fontsize=13, fontweight='bold')
plt.legend()
plt.tight_layout()
plt.show()
````

### 6.4 Logistic Growth (Nonlinear ODE)

$$\frac{du}{dt} = r\, u\left(1 - \frac{u}{K}\right), \quad u(0) = u_0$$

**Analytical solution (via separation of variables):**

$$u(t) = \frac{K}{1 + \left(\frac{K - u_0}{u_0}\right) e^{-rt}}$$

**Physical meaning:** Population growth with carrying capacity $K$

````python
# Parameters
r = 1.0    # growth rate
K = 100.0  # carrying capacity
t = np.linspace(0, 12, 400)

def logistic(t, u0, r=1.0, K=100.0):
    return K / (1 + ((K - u0) / u0) * np.exp(-r * t))

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Different initial conditions
for u0, color in zip([2, 10, 50, 120], ['purple', 'royalblue', 'green', 'tomato']):
    axes[0].plot(t, logistic(t, u0), color=color, lw=2, label=f'$u_0={u0}$')
axes[0].axhline(K, color='black', linestyle='--', alpha=0.6, label=f'Carrying capacity K={K}')
axes[0].set_xlabel('Time $t$')
axes[0].set_ylabel('Population $u(t)$')
axes[0].set_title('Logistic Growth (varying $u_0$)')
axes[0].legend(fontsize=9)

# Phase plot: du/dt vs u
u_vals = np.linspace(0, 140, 400)
dudt = r * u_vals * (1 - u_vals / K)
axes[1].plot(u_vals, dudt, 'darkgreen', lw=2.5)
axes[1].axhline(0, color='gray', linestyle='--', alpha=0.6)
axes[1].axvline(K, color='tomato', linestyle='--', alpha=0.6, label=f'K={K} (stable equilibrium)')
axes[1].axvline(0, color='purple', linestyle='--', alpha=0.6, label='u=0 (unstable equilibrium)')
axes[1].set_xlabel('$u$')
axes[1].set_ylabel('$du/dt$')
axes[1].set_title('Rate of Change vs Population')
axes[1].legend(fontsize=9)

plt.suptitle(r'Logistic Growth: $\frac{du}{dt} = ru\left(1 - \frac{u}{K}\right)$', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

## 7. Numerical Solutions

When no analytical solution exists, we use **numerical methods**. Understanding them helps you know what PINNs are replacing (or complementing).

### Common Methods:
| Method | Order | Notes |
|---|---|---|
| Euler | 1st | Simple, but inaccurate |
| Runge-Kutta 4 (RK4) | 4th | Gold standard for ODEs |
| `scipy.solve_ivp` | Adaptive | Uses RK45 by default |

Let's compare **analytical vs. Euler vs. RK4** on the SHO:

````python
# ---- Simple Harmonic Oscillator: u'' + omega^2 * u = 0
# Rewrite as system: y1 = u, y2 = u'
#   dy1/dt = y2
#   dy2/dt = -omega^2 * y1

omega = 2.0
dt = 0.05
T  = 4 * np.pi / omega
steps = int(T / dt)
t_num = np.linspace(0, T, steps)

def sho_rhs(t, y):
    """RHS of SHO as first-order system."""
    return [y[1], -omega**2 * y[0]]

y0 = [1.0, 0.0]  # [u(0), u'(0)]

# --- Euler Method ---
y_euler = np.zeros((steps, 2))
y_euler[0] = y0
for i in range(1, steps):
    dydt = sho_rhs(t_num[i-1], y_euler[i-1])
    y_euler[i] = y_euler[i-1] + dt * np.array(dydt)

# --- RK4 via scipy ---
sol_rk4 = solve_ivp(sho_rhs, [0, T], y0, method='RK45', t_eval=t_num, rtol=1e-8)

# --- Analytical ---
u_analytical = np.cos(omega * t_num)

# --- Plot ---
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].plot(t_num, u_analytical, 'black', lw=2.5, label='Analytical')
axes[0].plot(t_num, y_euler[:, 0], 'tomato', lw=1.8, linestyle='--', label=f'Euler (dt={dt})')
axes[0].plot(sol_rk4.t, sol_rk4.y[0], 'royalblue', lw=1.8, linestyle=':', label='RK45 (scipy)')
axes[0].set_xlabel('Time $t$')
axes[0].set_ylabel('$u(t)$')
axes[0].set_title('Numerical vs Analytical')
axes[0].legend()

# Error plot
euler_error = np.abs(y_euler[:, 0] - u_analytical)
rk4_error   = np.abs(sol_rk4.y[0] - u_analytical)

axes[1].semilogy(t_num, euler_error + 1e-16, 'tomato', lw=2, label='Euler error')
axes[1].semilogy(t_num, rk4_error  + 1e-16, 'royalblue', lw=2, label='RK45 error')
axes[1].set_xlabel('Time $t$')
axes[1].set_ylabel('|Error| (log scale)')
axes[1].set_title('Absolute Error vs Analytical')
axes[1].legend()

plt.suptitle('Numerical Methods Comparison: SHO', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print(f"Max Euler error: {euler_error.max():.6f}")
print(f"Max RK45 error:  {rk4_error.max():.2e}")
````

> 💡 **Observation:** Euler accumulates significant error, while RK45 stays very close to the analytical solution. A PINN should perform somewhere in between for simple problems, and better than grid-based methods for high-dimensional problems.

## 8. Why ODEs Matter for PINNs

PINNs embed the ODE **directly into the training loss**. Instead of discretizing time, the neural network $u_{\theta}(t)$ is trained so that:

$$\mathcal{L} = \underbrace{\frac{1}{N_r}\sum_{i=1}^{N_r} \left|\mathcal{F}[u_{\theta}](t_i)\right|^2}_{\text{Physics residual loss}} + \underbrace{\left|u_{\theta}(0) - u_0\right|^2}_{\text{IC loss}}$$

where $\mathcal{F}[u]$ is the ODE operator. For example:

- **Exponential decay:** $\mathcal{F}[u] = u' + ku$
- **SHO:** $\mathcal{F}[u] = u'' + \omega^2 u$
- **Logistic:** $\mathcal{F}[u] = u' - ru(1 - u/K)$

### Connection Table:

| ODE Concept | PINN Equivalent |
|---|---|
| ODE residual $\mathcal{F}[u] = 0$ | Physics loss term |
| Initial condition $u(0) = u_0$ | IC loss term |
| Boundary condition $u(L) = u_L$ | BC loss term |
| Domain $t \in [0, T]$ | Collocation point sampling region |
| Analytical solution $u(t)$ | Ground truth for error evaluation |

> 🔑 **The key insight:** Automatic differentiation computes $u'$, $u''$, etc., from the neural network — this is what makes PINNs possible!

````python
# Preview: how PyTorch computes derivatives automatically
# (Full PINN implementation is in notebook 03_pinn_basics/01_simple_ode.ipynb)

import torch

# Create a simple input tensor with gradient tracking
t_torch = torch.tensor([0.0, 0.5, 1.0, 1.5, 2.0], requires_grad=True)

# Suppose our "network" outputs cos(omega*t) — the analytical SHO solution
omega_t = torch.tensor(2.0)
u = torch.cos(omega_t * t_torch)

# Compute du/dt using autograd
du_dt = torch.autograd.grad(u.sum(), t_torch, create_graph=True)[0]

# Compute d²u/dt²
d2u_dt2 = torch.autograd.grad(du_dt.sum(), t_torch)[0]

print("t values:             ", t_torch.detach().numpy())
print("u = cos(2t):          ", u.detach().numpy().round(4))
print("du/dt = -2sin(2t):    ", du_dt.detach().numpy().round(4))
print("d²u/dt² = -4cos(2t):  ", d2u_dt2.detach().numpy().round(4))

# Verify SHO residual: d²u/dt² + omega^2 * u should be ≈ 0
residual = d2u_dt2 + omega_t**2 * u
print("\nSHO residual (should be ~0):", residual.detach().numpy().round(8))
````

> ✅ The residual is (numerically) zero — this is exactly what a trained PINN will enforce!
